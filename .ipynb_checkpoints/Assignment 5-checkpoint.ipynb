## 1. The Schwarzschild geometry
### a) (iii) A sketch of the shape of  light-like trajectory

Let's make $GM=1r$

$$W_{eff} (r) = \frac{1}{r^2} \left( 1 - \frac{2GM}{r} \right)$$

```python
M = 1
G = 1
r = np.arange(1.5, 15, 0.001)

def W_eff(r):
    W_eff = 1/(r**2) * (1 - (2*G*M)/r)
    return W_eff

plt.figure(figsize=(10, 4))
plt.plot(r, W_eff(r), label =r'$W_{eff}(r)$')
plt.plot([r[0], r[-1]], [W_eff(3*G*M), W_eff(3*G*M)], '--', label =r'$W_{eff}(3GM)$')
plt.grid()
# plt.yscale('symlog')

a1title = r'Radial effective potential for a light-like geodesic in the Schwarzschild geometry'
plt.title(a1title)
plt.xlabel(r'$r$ [GM]')
plt.ylabel(r'$W_{eff}(r)$ [J]')
plt.legend()
plt.show()
```

We can see from the graph and our calculations that $W_{eff}(3GM)$ is indeed a maximum. For light with an impact parameter of $\frac{1}{b^2} > W_{eff}(3GM)$, it will overcome the effective potential and get pulled into the spherical mass. It may exibit spiraling motion before impact if the ratio between the impact parameter and the radius of the spherical mass is greater than one. It is also dependent on angular momentum. 

## 2. Friedmann equation
### a) Matter dominated universe
$$1-\Omega (t) = \left( \frac{t}{t_0} \right) [1-\Omega (t_0)]$$

```python
t = 1
t0 = 14 * 365 * 24 * 3600 * 10**9
omega0 = 0.99

omega = 1-(1-omega0)*(t/t0)**(2/3)
print('{0}, {1:5e}, {2}'.format(omega, t0, -omega+1))
```
